#Decisão sobre os datos cuja data de lançamento está errada: Descartá-los

```python
#Criar um novo dataFrame sem os Relançamentos nas datas
df_novo2 = df[df['Data de lançamento'] != 'Relançamento'].copy()
```

```python
#Verificando os tipos das colunas
df_novo2.dtypes
```

```python
#COnverter a coluna data de lançamento para date_time
df_novo2['Data de lançamento'] = pd.to_datetime(df_novo2['Data de lançamento'])
```

```python
df_novo2.dtypes
```

```python
#Quais anos tivemos mais filmes lançados?
df_novo2['Data de lançamento'].dt.year.value_counts()
```

```python
df_novo2['Data de lançamento'].dt.year.value_counts().plot.bar()
```

```python
#Qual o filme com maior bilheteria?
df_novo2[df_novo2['Renda (R$) no ano de exibição'] == df_novo2['Renda (R$) no ano de exibição'].max()]
```

```python
#Outra maneira de exibir a maior bilheteria:
df_novo2.nlargest(1, 'Renda (R$) no ano de exibição')
```

```python
#Qual o filme com menor bilheteria?
df_novo2[df_novo2['Renda (R$) no ano de exibição'] == df_novo2['Renda (R$) no ano de exibição'].min()]
```
